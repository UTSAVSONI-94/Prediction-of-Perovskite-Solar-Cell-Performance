```python
import pandas as pd

file_path = 'data2.xlsx'
data = pd.read_excel(file_path)

data.head()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score, mean_squared_error

# Initialize the models
model1 = LinearRegression()  
model2 = LinearRegression()  
model3 = LinearRegression()  
model4 = LinearRegression()  

# Fit the models
model1.fit(data[['x1']], data['y1'])  # Assuming 'y' is your target variable
model2.fit(data[['x2']], data['y2'])
model3.fit(data[['x3']], data['y3'])
model4.fit(data[['x4']], data['y4'])

# Predict using the fitted models
y1_pred = model1.predict(data[['x1']])
y2_pred = model2.predict(data[['x2']])
y3_pred = model3.predict(data[['x3']])
y4_pred = model4.predict(data[['x4']])

# You can now evaluate the predictions, for example, with r2_score or mean_squared_error
```

```python
def predict_all(x_val):
    # Prepare the input as a 2D array for prediction
    x1_val = np.array([[x_val]])
    x2_val = np.array([[x_val]])
    x3_val = np.array([[x_val]])
    x4_val = np.array([[x_val]])
    
    # Predict for each model
    y1_pred = model1.predict(x1_val)[0]
    y2_pred = model2.predict(x2_val)[0]
    y3_pred = model3.predict(x3_val)[0]
    y4_pred = model4.predict(x4_val)[0]
    
    return y1_pred, y2_pred, y3_pred, y4_pred
# Example usage:
x_val = 111.157025 
y1_pred, y2_pred, y3_pred, y4_pred = predict_all(x_val)

# Output the results
print(f"For x = {x_val}:")
print(f"Predicted y1 = {y1_pred}")
print(f"Predicted y2 = {y2_pred}")
print(f"Predicted y3 = {y3_pred}")
print(f"Predicted y4 = {y4_pred}")
```
